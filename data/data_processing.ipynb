# Processing the data

## Drop unnecessary columns

For our case. The Columns
* RowNumber
* CustomerId
* Surname

Are not necessary for the analysis. The 3 columns are follow each other.

```python
raw_dataset_churn = raw_dataset_churn.drop(raw_dataset_churn.columns[0:3], axis = 1)
raw_dataset_churn
```

## Transform categorical data into a numerical representation

```python
label_encoder = LabelEncoder()
```

### Geography

```python
raw_dataset_churn['Geography'] = label_encoder.fit_transform(raw_dataset_churn['Geography'])
label_encoder.classes_
```

```python
raw_dataset_churn
```

### Gender

```python
raw_dataset_churn['Gender'] = label_encoder.fit_transform(raw_dataset_churn['Gender'])
label_encoder.classes_
```

```python
raw_dataset_churn
```

## Normalize the data

```python
raw_dataset_churn.iloc[:, :-1]
```

```python
scaler = StandardScaler()
raw_dataset_churn_normalized = scaler.fit_transform(raw_dataset_churn.iloc[:, :-1])
raw_dataset_churn_normalized = pd.DataFrame(raw_dataset_churn_normalized, columns = raw_dataset_churn.iloc[:, :-1].columns)
raw_dataset_churn_normalized['Exited'] = raw_dataset_churn.iloc[:, -1]
raw_dataset_churn_normalized
```
